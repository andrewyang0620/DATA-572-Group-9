```python
import pandas as pd
df = pd.read_csv("titanic_augmented.csv")

cols_to_drop = [
    "PassengerId",
    "Name",
    "Ticket",
    "Cabin",
    "booking_reference",
    "service_id"
]

df_clean = df.drop(columns=cols_to_drop)

df_clean.head()
```

```python
from sklearn.model_selection import train_test_split

y = df_clean["Survived"]
X = df_clean.drop(columns=["Survived"])

X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

train_raw = X_train_raw.copy()
train_raw["Survived"] = y_train.values

test_raw = X_test_raw.copy()
test_raw["Survived"] = y_test.values

train_raw.to_csv("train_raw.csv", index=False)
test_raw.to_csv("test_raw.csv", index=False)

print("Overall:", y.mean())
print("Train  :", y_train.mean())
print("Test   :", y_test.mean())
print("Shapes :", X_train_raw.shape, X_test_raw.shape)
```
